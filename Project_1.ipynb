```python

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(" " , "_")
df.columns
```

```python
df = df.rename(columns = {
    'purchase_amount_(usd)' : 'purchase_amount'
})
```

```python
df.columns
```

```python
#Create a purchase_frequency_days column 
print(df['frequency_of_purchases'].unique())
```

```python
frequency_days = {
    'Fortnightly' : 14 ,
    'Weekly' : 7 ,
    'Annually' : 360,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Monthly': 30,
    'Every 3 Months':90 
}
df['frequency_of_days'] = df['frequency_of_purchases'].map(frequency_days)
```

```python
#Create a column of age_group
labels = ['Young Adult' , 'Adult' , 'Middle-Aged' , 'Senior']
df['age_group'] = pd.qcut(df['age'] , q=4 , labels = labels)
```
